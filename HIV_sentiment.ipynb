```python
def clean_article(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
```

```python
def analyze_sentiment(tweet):
        analysis = TextBlob(clean_article(tweet))
        
        if analysis.sentiment.polarity > 0:
            return 1
        elif analysis.sentiment.polarity == 0:
            return 0
        else:
            return -1
```

```python
def analyze_subjectivity(tweet):
        analysis = TextBlob(clean_article(tweet))
        return analysis.sentiment.subjectivity
#         if analysis.sentiment.subjectivity > 0.5:
#             return 1
#         else:
#             return 0
```

```python
data['sentiment'] = np.array([analyze_sentiment(art) for art in data['name']])
```

```python
data['subjectivity'] = np.array([analyze_subjectivity(art) for art in data['name']])
```
